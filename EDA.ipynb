```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('/content/drive/MyDrive/NLP PROJECT/data.csv')
df.head()
```

```python
df.shape
```

```python
# Dropping Duplicates

df=df.drop_duplicates(subset=['SEQUENCE', 'LABEL'], keep='first')
df.shape
```

# Exploratory Data Analysis

```python
# Visualize the distribution of labels
label_distribution = df['LABEL'].value_counts()
fig = px.bar( x=label_distribution.index, y=label_distribution.values, labels={'x': 'Label', 'y': 'Count'},
             title='Distribution of Labels')
fig.update_layout(width=800)
fig.show()
```

```python
organism_train = df['CREATURE'].value_counts()
fig = px.bar(x=organism_train.index, y=organism_train.values,
             title='Distribution of Organisms')
fig.update_layout(width=800)
fig.show()
```

```python
df['Sequence Length'] = df['SEQUENCE'].apply(len)

fig = px.histogram(df, x='Sequence Length', nbins=20, title='Distribution of Sequence Lengths ')
fig.update_layout(width=800)
fig.show()
```

```python
from collections import Counter

amino_acid_codes = ''.join(df['SEQUENCE']) # Concatenate all sequences

# Count the frequency of each amino acid code
code_frequency = Counter(amino_acid_codes)
```

```python
df_frequency = pd.DataFrame(list(code_frequency.items()), columns=['Amino Acid Code', 'Frequency'])
df_frequency = df_frequency.sort_values(by='Frequency', ascending=False)
print('Total unique codes in train set:', df_frequency.shape[0])

# Plot the frequency
fig_frequency = px.bar(df_frequency, x='Amino Acid Code', y='Frequency', title='Amino Acid Code Frequencies')
fig_frequency.show()
```

amino acids (X, U, B, Z) are present in very less quantity.

# Splitting the dataset

 I have segmented the data into two distinct sets. The test_set comprises rows where the 'CREATURE' column takes values 'creature0' or 'creature2', while the train_set includes rows where the 'CREATURE' column does not have the values 'creature0' or 'creature2'.

```python
# Splitting the Data

test_set = df[df['CREATURE'].isin(['creature0','creature2'])]
train_set = df[~df['CREATURE'].isin(['creature0','creature2'])]

test_set.shape, train_set.shape
```
